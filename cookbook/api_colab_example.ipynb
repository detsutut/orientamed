```
# @title Query Parameters { display-mode: "form" }

text = 'il paziente lamenta dolore articolare, in particolare agli arti inferiori' # @param {type:"string"}
# @markdown ---
# @markdown Options
overlap = 100 # @param {type:"slider", min:0, max:100, step:1}
filter_tags = [] # @param {"type":"raw","placeholder":"[]"}
exclude = False # @param {type:"boolean"}
premium = False # @param {type:"boolean"}
```

Request using `GET`

```
import requests
import pandas as pd
url = "https://dheal-com.unipv.it:7878/extract"
params = {'text': text, 'f': filter_tags,'e': exclude,'o': overlap,'p': premium}
response = requests.get(url, params=params)
pd.DataFrame(response.json())
```

```
import requests
import pandas as pd
import json
url = "https://dheal-com.unipv.it:7878/extract"
data = {'text': text, 'filter_tags': filter_tags,'exclude': exclude,'fuzzy_threshold': overlap,'use_premium': premium}
response = requests.post(url, data=json.dumps(data), headers={"ContentType": "application/json"})
pd.DataFrame(response.json())
```
